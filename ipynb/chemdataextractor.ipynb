```python
from chemdataextractor import Document
import pandas as pd
import os
```

```python
chemical_list_fr = []
for filename in os.listdir('./flame_retardants/'):
    if filename.endswith(".html"): 
        print(filename)
        f = open('./flame_retardants/'+filename,'rb')
        doc = Document.from_file(f)
        for i in range(len(doc.cems)):
            chemical_list_fr.append(str(doc.cems[i]))
```

```python
counts = Counter(chemical_list_fr)
type(counts)
flame_retardant_chemical_list = pd.DataFrame.from_dict(counts, orient='index').sort_values(by = [0], ascending=False)
flame_retardant_chemical_list.to_csv('./flame_retardant_chemical.csv')
```
